### 2nd block

```python
# creates dataframe from csv file
data = pd.read_csv('problem_set_1_data.csv')

# changes column names LOOK INTO THIS SHIT AGAIN
data = data.rename(columns={'quizzes':'quiz_total', 'attendedClass':'attended',
                     'finalgrade':'final_grade', 'final_EXAM':'final_exam',
                     'student':'name'})

print('Data head:', '\n', data.head())
print()
print('Data tail:', '\n', data.tail())
print()
print('Data description:', '\n', data.describe())
```

The average of quizzes means that the average quiz score was a 69.13% for the class. The average for amount of people attended was .39, meaning very little people attended lecture. The mean for the final grade is a 91.86%, meaning that, for the most part, everyone did very well in the class. The mean for the final exam is 39.67. I'm assuming that this represents the average amount of points people got rather than the average percent for the final exam.

### 3rd block

```python
# creates dataframe with only students who attended
attended_data = data[data['attended'] == 1]

# creates dataframe with only students who did not attend
unattended_data = data[data['attended'] == 0]

print('Attended data:', '\n', attended_data.mean())
print()
print('Unattended data:', '\n', unattended_data.mean())
```

Based off of the average scores for 'quiz_total', 'final_grade', and 'final_exam', the comparison between students who attended lectures versus not attending lecture show that students who attended lectures have a higher average on all categories.
