### 1st block (Import Libraries)

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

### 2nd block (Creating Function)

```python
def weather(zipcode):

  # creates a complete url 
  url = 'https://weather.com/weather/today/l/' + str(zipcode) + ':4:US'

  # retrieves the HTML from code
  response = requests.get(url)

  response_status = response.status_code

  # checks if connection to weather website was successful
  if response_status != 200:
    print('The zipcode entered is not valid. \n Please try again and enter a valid 5 digit zipcode')
  
  else:

    # creates html soup 
    soup = BeautifulSoup(response.content)

    # retrieves location of the entered zipcode
    area = soup.find(class_ = 'CurrentConditions--location--1Ayv3').text

    # retrieves the temperature of the entered zipcode
    temp_data = soup.find(class_ = 'CurrentConditions--tempValue--3KcTQ').text
    # specificies temperature measurement
    temp = temp_data + ' Fahrenheit'
    
    # prints area, zipcode, and temperature
    print(area)
    print(zipcode)
    print(temp)
```

### 3rd block (Passing in Zipcodes)
