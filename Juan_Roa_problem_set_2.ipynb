### 1st block (importing libraries)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### 2nd block (creating figure)

```python
anscombe = sns.load_dataset('anscombe')
data1 = anscombe[anscombe['dataset'] ==  'I']
data2 = anscombe[anscombe['dataset'] ==  'II']
data3 = anscombe[anscombe['dataset'] ==  'III']
data4 = anscombe[anscombe['dataset'] ==  'IV']
```

```python
data1_changed = data1.drop(['dataset'], axis=1)
data1_changed

datasets = [data1, data2, data3, data4]
cleaned_data = []

for i in datasets:
  i = i.drop(['dataset'], axis=1)
  cleaned_data.append(i)
```

```python
means = []
stds = []

for i in range(len(cleaned_data)):
  mean_data = str(cleaned_data[i].mean()).replace('dtype: float64', '')
  means.append(mean_data)
  std_data = str(cleaned_data[i].std()).replace('dtype: float64', '')
  stds.append(std_data)

print(means, '\n', stds)
```

```python
mean_text = []
std_text = []

for i in range(len(means)):
  mean = 'Mean: \n' + means[i]
  std = 'Standard Dev: \n' + stds[i]
  mean_text.append(mean)
  std_text.append(std)

print(mean_text, '\n', std_text)
```

```python
mean_std = []
for i in range(4):
  text = mean_text[i] + std_text[i]
  mean_std.append(text)

mean_std
```

```python
fig, axes = plt.subplots(2,2, figsize=[10,10])


for i in np.arange(0, axes.shape[0]):
  for j in np.arange(0, axes.shape[1]):

    axes[i,j].set_xlabel('x')
    axes[i,j].set_ylabel('y')
    axes[i,j].set_xlim(0, 20)
    axes[i,j].set_ylim(0, 14)
    axes[i,j].set_xticks(np.arange(0, 21, 4))
    axes[i,j].set_yticks(np.arange(0, 15, 2))

    if i == 0 and j == 0:
      axes[i,j].scatter(cleaned_data[0]['x'], cleaned_data[0]['y'], c='green')
      axes[i,j].set_title('Dataset I')
      axes[i,j].text(0.5, 9.5, s=mean_std[0])

    elif i == 0 and j == 1:
      axes[i,j].scatter(cleaned_data[1]['x'], cleaned_data[1]['y'], c='orange')
      axes[i,j].set_title('Dataset II')
      axes[i,j].text(0.5, 9.5, s=mean_std[1])

    elif i == 1 and j == 0:
      axes[i,j].scatter(cleaned_data[2]['x'], cleaned_data[2]['y'], c='blue')
      axes[i,j].set_title('Dataset III')
      axes[i,j].text(0.5, 9.5, s=mean_std[2])

    elif i == 1 and j == 1:
      axes[i,j].scatter(cleaned_data[3]['x'], cleaned_data[3]['y'], c='red')
      axes[i,j].set_title('Dataset IV')
      axes[i,j].text(0.5, 9.5, s=mean_std[3])



#axes[0,0].scatter(cleaned_data[0]['x'], cleaned_data[0]['y'], c='orange')
```
