### 1st block (importing libraries)

```python
import pandas as pd
```

### 2nd block (web scraping)

```python
url = 'https://en.wikipedia.org/wiki/MonsterVerse'
data = pd.read_html(url)
df = data[3]
```

```python
df.head()
```

### 3rd block (reorganizing data)

```python
df.index = df[df.columns[0]]
df = df.drop(columns=df.columns[0])
df.columns = ['Godzilla', 'Kong:Skull Island',
              'Godzilla:King of the Monsters', 'Godzilla vs. Kong']
df.index.name = 'character'
df = df.drop(['Humans'])
```
